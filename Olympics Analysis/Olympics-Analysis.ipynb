```python
data.drop_duplicates(inplace=True)
```

```python
sns.countplot(data=data,x="Sex")
plt.title("Distribution by Gender")
plt.show()
```

```python
sns.histplot(data=data,x="Age",bins=20,kde=True)
plt.title("Age Distribution")
plt.show()
```

```python
sns.histplot(data=data,x="Height",bins=20,kde=True)
plt.title("Height Distribution")
plt.show()
```

```python
sns.histplot(data=data,x="Weight",bins=20,kde=True)
plt.title("Weight Distribution")
plt.show()
```

```python
sns.countplot(data=data,x="Medal")
plt.title("Medal  Distribution")
plt.show()
```

```python
sns.countplot(data=data,x="Year",hue="Medal")
plt.title("Medal  Distribution over the years")
plt.xticks(rotation=45)
plt.show()
```

```python
years_avg=data.groupby("Year")["Age"].mean()
print(years_avg)
```

```python
sport_med_height=data.groupby("Sport")["Height"].median()
print(sport_med_height)
```

```python
country_gender=data.groupby(["NOC","Sex"])["ID"].count()
print(country_gender)
```

```python
country_gold=data[data["Medal"]=="Gold"].groupby("NOC")["Medal"].count()
print(country_gold)
```

```python
country_gold[country_gold==country_gold.max()]
```

```python
sport_gender_avg_weight = data.groupby(["Sport", "Sex"])["Weight"].mean()
print(sport_gender_avg_weight)
```

```python
sport_event_count=data.groupby("Sport")["Event"].nunique().sort_values(ascending=False)
sport_event_count.plot(kind="bar")
plt.title("No. of Unique Events per Sport")
plt.xlabel("Sport")
plt.ylabel("Number of Unique Events")
plt.xticks(rotation=45)
plt.show()
```

```python
year_participation=data.groupby("Year")["ID"].nunique()
year_participation.plot(kind="bar")
plt.title("No. of Participants per year")
plt.xlabel("Year")
plt.ylabel("Number of Participants")
plt.show()
```

```python
country_avg_age=data.groupby("NOC")["Age"].mean().sort_values(ascending=False)
country_avg_age.head(10).plot(kind="bar")
plt.title("Top 10 countries with highest average age")
plt.xlabel("Country")
plt.ylabel("Average Age")
plt.xticks(rotation=45)
plt.show()
```

```python
sns.violinplot(data=data,x="Medal",y="Height",palette="Set2")
plt.title("Distribution of Heights by Medals")
plt.xlabel("Medal")
plt.ylabel("Height")
plt.show()
```

```python
sns.scatterplot(data=data,x="Height",y="Weight",hue="Medal",palette="Set1")
plt.title("Athelte Height vs Weight by Medal Status")
plt.xlabel("Height")
plt.ylabel("Weight")
plt.legend(title="Medal")
plt.show()
```
