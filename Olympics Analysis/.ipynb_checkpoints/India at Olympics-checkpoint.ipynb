```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
file_path = r'C:\Users\KIIT\Desktop\EDA Projects\Olympics Analysis\archive (1)\dataset_olympics.csv'
data = pd.read_csv(file_path)
```

```python
india_data = data[data['Team'] == 'India']
```

```python
participants_by_year = india_data.groupby('Year')['ID'].nunique()
participants_by_year.plot(kind='bar', figsize=(12, 6), color='skyblue')
plt.title('Number of Participants from India by Year')
plt.xlabel('Year')
plt.ylabel('Number of Participants')
plt.show()
```

```python
medals_by_year = india_data.dropna(subset=['Medal']).groupby('Year')['Medal'].count()
medals_by_year.plot(kind='bar', figsize=(12, 6), color='green')
plt.title('Number of Medals Won by India by Year')
plt.xlabel('Year')
plt.ylabel('Number of Medals')
plt.show()
```

```python
medals_by_sport = india_data.dropna(subset=['Medal']).groupby('Sport')['Medal'].count().sort_values()
medals_by_sport.plot(kind='barh', figsize=(12, 8), color='orange')
plt.title('Number of Medals Won by India by Sport')
plt.xlabel('Number of Medals')
plt.ylabel('Sport')
plt.show()
```

```python
medals_by_type = india_data['Medal'].dropna().value_counts()
medals_by_type.plot(kind='pie', autopct='%1.1f%%', figsize=(8, 8), colors=['gold', 'silver', 'brown'])
plt.title('Distribution of Medals Won by India by Type')
plt.ylabel('')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.histplot(india_data['Age'].dropna(), bins=20, kde=True)
plt.title('Age Distribution of Indian Athletes')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.histplot(india_data['Height'].dropna(), bins=20, kde=True, color='blue')
plt.title('Height Distribution of Indian Athletes')
plt.xlabel('Height (cm)')
plt.ylabel('Frequency')
plt.show()

plt.figure(figsize=(12, 6))
sns.histplot(india_data['Weight'].dropna(), bins=20, kde=True, color='red')
plt.title('Weight Distribution of Indian Athletes')
plt.xlabel('Weight (kg)')
plt.ylabel('Frequency')
plt.show()
```

```python
medals_by_gender = india_data.dropna(subset=['Medal']).groupby('Sex')['Medal'].count()
medals_by_gender.plot(kind='bar', figsize=(8, 6), color=['blue', 'pink'])
plt.title('Medals Won by Indian Athletes by Gender')
plt.xlabel('Gender')
plt.ylabel('Number of Medals')
plt.xticks(ticks=[0, 1], labels=['Male', 'Female'])
plt.show()
```

```python
top_medalists = india_data.dropna(subset=['Medal']).groupby('Name')['Medal'].count().sort_values(ascending=False).head(10)
top_medalists.plot(kind='barh', figsize=(12, 8), color='purple')
plt.title('Top Medalists from India')
plt.xlabel('Number of Medals')
plt.ylabel('Athlete')
plt.show()
```

```python
fig, ax1 = plt.subplots(figsize=(12, 6))

ax2 = ax1.twinx()
ax1.plot(participants_by_year.index, participants_by_year.values, 'g-')
ax2.plot(medals_by_year.index, medals_by_year.values, 'b-')

ax1.set_xlabel('Year')
ax1.set_ylabel('Number of Participants', color='g')
ax2.set_ylabel('Number of Medals', color='b')
plt.title('Trends Over Time in Terms of Participation and Success for India')
plt.show()
```
