```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("Netflix Userbase.csv")
```

```python
average_rev = df.groupby("Subscription Type")["Monthly Revenue"].mean()
```

```python
user_count=df.groupby("Country")["User ID"].count()
```

```python
print(user_count)
```

```python
print(average_rev)
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age', y='Monthly Revenue', hue='Subscription Type', data=df)
plt.title('Age vs Monthly Revenue')
plt.xlabel('Age')
plt.ylabel('Monthly Revenue')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.pairplot(df, vars=['Monthly Revenue', 'Age'], hue='Device')
plt.title('Pairplot of Monthly Revenue and Age by Device')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Subscription Type', data=df)
plt.title('Subscription Type Distribution')
plt.xlabel('Subscription Type')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Subscription Type', y='Monthly Revenue', data=df)
plt.title('Monthly Revenue Distribution by Subscription Type')
plt.xlabel('Subscription Type')
plt.ylabel('Monthly Revenue')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Subscription Type', y='Age', data=df)
plt.title('Age Distribution by Subscription Type')
plt.xlabel('Subscription Type')
plt.ylabel('Age')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(y='Country', data=df, order=df['Country'].value_counts().index)
plt.title('Country Distribution')
plt.xlabel('Count')
plt.ylabel('Country')
plt.show()
```
